```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prompts = ("prompt 1 (1)", "prompt 1 (2)", "prompt 2", "prompt 4")
colors = ['#43aa8b', '#f94144', '#f8961e']
```

```python
company_validations = {
    'Right': (68, 65, 67, 70),
    'Wrong': (14, 16, 17, 14),
    'Not sure': (18, 19, 16, 16),
}

x = np.arange(len(prompts))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

color_pick = 0

for attribute, measurement in company_validations.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=colors[color_pick])
    ax.bar_label(rects, padding=3)
    multiplier += 1
    color_pick += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Validation')
ax.set_title('Prompts')
ax.set_xticks(x + width, prompts)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 100)
ax.set_title('Validation on Company')

plt.show()
```
