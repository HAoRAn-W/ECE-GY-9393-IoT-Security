```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = [i for i in range(1, 19)]
y_pos = np.arange(len(people))
performance = [395.019154548645, 382.78102684020996, 373.190536737442, 377.93091011047363, 383.13956475257874, 378.00438952445984, 391.964370012283, 393.3709087371826, 392.54962682724, 394.5913391113281,387.5093297958374, 384.8831992149353, 393.15416526794434, 390.0428512096405, 410.6316804885864, 385.0282280445099, 381.20925068855286, 308.0199885368347]
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center', color='#78b9d2')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Time (seconds)')
ax.set_title('Time Used per 200 Queries')
ax.axvline(np.mean(performance), linestyle='--', color='black')

plt.show()
```
